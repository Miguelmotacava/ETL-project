```python
import pandas as pd
df_final2 = pd.read_csv('scrap_players2.csv', sep = ' ')
df_final = pd.read_csv('scrap_players.csv', sep = ' ')
```

```python
dic={}
keys = list(df_final2['names'])
values = list(df_final2['equipos'])
for i in range(len(keys)):
    dic[keys[i]]=values[i]
```

```python
team = {}
for i in dic.keys():
    for j in df_final.name_player:
        if j in i:
            team[j]=dic[i]
        else:
            pass
```

```python
team_player=[]
for ind, val in enumerate(df_final.name_player):
    if val in list(team.keys()):
        team_player.append(team[val])
    else:
        team_player.append('None')
        
```

```python
df_final['team_player']=team_player
```

```python
df_final=df_final[['name_player', 'team_player', 'media', 'posicion', 'tipo', 'precio',
       'skill', 'wf', 'ritmo', 'shot', 'pas', 'drib', 'def', 'phy',
       'popularity', 'base stats', 'igs', 'plataforma', 'time']]
```
